```python
df['Occupation']=df['Occupation'].fillna(df['Occupation'].mode()[0])
df['Relationship']=df['Relationship'].fillna(df['Relationship'].mode()[0])
df['Race']=df['Race'].fillna(df['Race'].mode()[0])
df['Sex']=df['Sex'].fillna(df['Sex'].mode()[0])
df['Native_country']=df['Native_country'].fillna(df['Native_country'].mode()[0])
df['Income']=df['Income'].fillna(df['Income'].mode()[0])
```

```python
df['Capital_gain']=df['Capital_gain'].fillna(df['Capital_gain'].mean())
df['Capital_loss']=df['Capital_loss'].fillna(df['Capital_loss'].mean())
df['Hours_per_week']=df['Hours_per_week'].fillna(df['Hours_per_week'].mean())
```

# Training the model

```python
from sklearn.model_selection import cross_val_score
def classify(model,x,y):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
    model.fit(x_train,y_train)
    print("Accuracy",model.score(x_test,y_test)*100)
    score=cross_val_score(model,x,y,cv=5)
    print("Cross Validation",np.mean(score)*100)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
classify(model,x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
classify(model,x,y)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
classify(model,x,y)
```
